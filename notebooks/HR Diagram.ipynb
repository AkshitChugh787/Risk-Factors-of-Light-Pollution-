```python
O = df[df['Spectral Class'] == 'O']
B = df[df['Spectral Class'] == 'B']
A = df[df['Spectral Class'] == 'A']
F = df[df['Spectral Class'] == 'F']
G = df[df['Spectral Class'] == 'G']
K = df[df['Spectral Class'] == 'K']
M = df[df['Spectral Class'] == 'M']
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mplcyberpunk

plt.style.use('cyberpunk')
```

```python
plt.figure(figsize=(7,10))
plt.scatter(np.log(O["Temperature (K)"]), np.log(O["Luminosity(L/Lo)"]),100,c='#0526ba' ,label = 'O')
plt.scatter(np.log(B["Temperature (K)"]), np.log(B["Luminosity(L/Lo)"]),100,c ='#0d8fc6', label = 'B')
plt.scatter(np.log(A["Temperature (K)"]), np.log(A["Luminosity(L/Lo)"]),100,c='#FFF', label = 'A')
plt.scatter(np.log(F["Temperature (K)"]), np.log(F["Luminosity(L/Lo)"]),100,c='#f4e97f', label = 'F')
plt.scatter(np.log(G["Temperature (K)"]), np.log(G["Luminosity(L/Lo)"]),100,c='#ffe605', label = 'G')
plt.scatter(np.log(K["Temperature (K)"]), np.log(K["Luminosity(L/Lo)"]),100,c='#f50' ,label = 'K')
plt.scatter(np.log(M["Temperature (K)"]), np.log(M["Luminosity(L/Lo)"]),100,c='#b30000', label = 'M')
plt.gca().invert_xaxis()
plt.title("Hertzsprung - Russell Diagram")
plt.ylabel("Luminosity")
plt.xlabel("Temperature")
plt.legend()
```
